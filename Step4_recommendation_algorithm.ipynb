```python
buz_dict = {}
for buz_id in all_business:
    print(buz_id)
    dish_dict = {}
    dish_candidate = set(expanded_data.loc[expanded_data['business_id'] == buz_id, 'expanded_DishInText'])
    for dish in dish_candidate:
        score = expanded_data.loc[(expanded_data['business_id'] == buz_id) 
                                  & (expanded_data['expanded_DishInText'] == dish), 'weighted_score']
        dish_dict[dish] = (score.mean(), score.sum())
    buz_dict[buz_id] = dish_dict
```
