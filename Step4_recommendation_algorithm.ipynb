```python
for idx in range(len(data)):
    temp = set((re.findall(r"'(.*?)'", data.at[idx, 'DishInText'])))
    while temp:
        dish = temp.pop()
        temp_dict = {'business_id': data.at[idx,'business_id'], 
                     'weighted_score': data.at[idx,'weighted_score'], 
                     'DishInText': dish}
        new_data.append(temp_dict, ignore_index = True)
    if idx % 10000 == 0:
        print(idx)
```

```python
new_data = new_data.loc[2: , ['business_id', 'weighted_score', 'DishInText']]
```

```python
buz_dict = {}
for buz_id in all_business:
    dish_dict = {}
    dish_candidate = set(data.loc[data['business_id'] == buz_id, 'DishInText'])
    for dish in dish_candidate:
        score = data.loc[(data['business_id'] == buz_id) & (data['DishInText'] == dish), 'weighted_score']
        dish_dict[dish] = (score.mean(), score.sum())
    buz_dict[buz_id] = dish_dict
```
