```python
restaurants = []
for item in df["categories"].unique().tolist():
    if isinstance(item, str):
        if ("Restaurants" in item) or ("Food" in item):
            restaurants.append(item)    
```

```python
restaurant_id = []
for idx, item in df.iterrows():
    if (isinstance(item["categories"], str)):
        if ("Restaurants" in item["categories"]) or ("Food" in item["categories"]):
            restaurant_id.append(item["business_id"])
```

```python
restaurant_id = set(restaurant_id)
```
