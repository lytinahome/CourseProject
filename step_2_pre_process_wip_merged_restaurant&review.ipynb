```python
import numpy as np
import pandas as pd
df_business = pd.read_csv("./business.csv")

df_business.columns.tolist()
```

```python

test_busi = df_business.at[0,"business_id"]
print(test_busi)

```

```python
df_review = pd.read_csv("review.csv")
```

```python
idx = []
for i in range(len(df_review)):
    if i%10000 == 0:
        print(i)
    if test_busi == df_review.at[i, "business_id"]:
        idx.append(i)
print(idx)
```

```python
restaurant_id = []
name = []
for idx, item in df_business.iterrows():
    if (isinstance(item["categories"], str)):
        if ("Restaurants" in item["categories"]) or ("Food" in item["categories"]):
            restaurant_id.append(item["business_id"])
            name.append(item["name"])
df_restaurant = pd.DataFrame(zip(restaurant_id, name), columns = ["business_id", "name"])
df_restaurant.to_csv("./restaurant.csv",index=False)
```

```python
df_merge_restaurant_review = pd.merge(df_restaurant, df_review, on="business_id")
```
