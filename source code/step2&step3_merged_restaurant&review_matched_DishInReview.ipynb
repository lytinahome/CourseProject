```python
import numpy as np
import pandas as pd
df_business = pd.read_csv("./business.csv")

df_business.columns.tolist()
```

```python
df_review = pd.read_csv("review.csv")
```

```python
restaurant_id = []
name = []
for idx, item in df_business.iterrows():
    if (isinstance(item["categories"], str)):
        if ("Restaurants" in item["categories"]) or ("Food" in item["categories"]):
            restaurant_id.append(item["business_id"])
            name.append(item["name"])
df_restaurant = pd.DataFrame(zip(restaurant_id, name), columns = ["business_id", "name"])
df_restaurant.to_csv("./restaurant.csv",index=False)
```

```python
len(df_restaurant), len(df_business)
```

```python
df_merge_restaurant_review = pd.merge(df_restaurant, df_review, on="business_id")
```

```python
dish_1 = pd.read_excel("./List of food in AMER and EU.xlsx")
dish_2 = pd.read_excel("./CS410 rest of world dishes.xlsx")
dish_1.columns = dish_2.columns
dishlist = pd.concat([dish_1, dish_2], ignore_index = True)
dishlist.to_csv("./all_dish.csv",index=False)
```

```python
import re
set_dishlist1 = set((list(dishlist["Dish"].str.lower()))) #each row in dishlist is considered a dish
set_dishlist2 = set(re.findall(r'\w+', str(list(dishlist["Dish"].str.lower())))) #each word in each row is considered a  dish
#set_dishlist = list(set_dishlist1.union(set_dishlist2))
set_dishlist = set_dishlist1
print(set_dishlist, "\n", len(set_dishlist), len(set_dishlist1), len(set_dishlist2))
```

```python
#DishInText = lambda x : set(re.findall(r'\w+', x.str.lower()))&set_dishlist
#df_merge_restaurant_review["DishInText"] = df_merge_restaurant_review["text"].apply(DishInText)
df_merge_restaurant_review["DishInText"] = None
#df_merge_restaurant_review["Mark"] = None
print(df_merge_restaurant_review["text"])
```

```python
def DishInText(dishlist:list, Text:str):
    common = []
    #words = set(re.findall(r'\w+', Text.lower()))
    for i in dishlist:
        if i in Text:
            common.append(i)
    return set(common)
```

```python
for i in range(len(df_merge_restaurant_review)):
    temp = DishInText(set_dishlist, df_merge_restaurant_review.at[i, "text"].lower())
    if len(temp)>0:
        df_merge_restaurant_review.at[i, "DishInText"] = temp
        #df_merge_restaurant_review["Mark"] = 1
    if i%10000 == 0: print(i)
```

```python
#keep data where dishes were mentioned and get the corresponding ratings
#columns kept include: "business_id", "stars","useful","funny", "cool", "DishInText"
df_notext = df_merge_restaurant_review[df_merge_restaurant_review["DishInText"].apply(lambda x: x is not None)][["business_id", "stars","useful","funny", "cool", "DishInText"]]
```
